```python
import pandas as pd
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn import metrics
```

```python
dataset = pd.read_csv("PlayTennis.csv")
print(dataset)
```

```python
all_cols=dataset.columns
print(type(all_cols))
```

```python
feature_cols=all_cols[1:5]
print(feature_cols)
```

```python
#feature encoding
for label in dataset.columns:
    dataset[label]=LabelEncoder().fit_transform(dataset[label])
    print(dataset[label])
```

```python
#formatting into numpy array
data=dataset.drop(['Play Tennis'],axis=1)
print(data)

target=np.array(dataset['Play Tennis'])
print(target)
```

```python
#splitting the dataset
# X_train,X_test,y_train,y_test=train_test_split(data,target,test_size=0.3)
X_train,X_test,y_train,y_test=train_test_split(data,target,test_size=0.3,shuffle=False,random_state=42)
print(y_test)
```

```python
#Train Desicion tree classifier
id3=DecisionTreeClassifier(criterion="entropy")
id3=id3.fit(X_train,y_train)
print(id3)
```

```python
#predict the response for test dataset
y_pred=id3.predict(X_test)
print("Actual output:",y_test)
print("predicted output:",y_pred)
print("Accuracy:",metrics.accuracy_score(y_test,y_pred))
```
