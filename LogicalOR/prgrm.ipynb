```python
import numpy as np
```

```python
class Perceptron(object):

    """Implements a perceptron network"""

    def __init__(self, input_size, lr=1, epochs=5):

        self.W = np.zeros(input_size)

        self.b=0

        # add one for bias

        self.epochs = epochs

        self.lr = lr

   

    def activation_fn(self, x):

        #return (x >= 0).astype(np.float32)

        return 1 if x > 0 else 0

 

    def predict(self, x):

        z = self.W.T.dot(x)

        a = self.activation_fn(z+self.b)

        return a

 

    def fit(self, X, d):

       

        for _ in range(self.epochs):

            for i in range(d.shape[0]):

                #x = np.insert(X[i], 0,1)

                x=X[i]

                y = self.predict(x)

                e = d[i] - y

                self.W = self.W + self.lr * e * x

                self.b=self.b + e

                print(self.b)

   

       
```

```python
 #OR gate              

if __name__ == '__main__':

    Input_OR = np.array([

        [0, 0],

        [0, 1],

        [1, 0], 

        [1, 1]

    ])

    target_val = np.array([0, 1, 1, 1])

 

  

    perceptron_OR = Perceptron(input_size=2)

    perceptron_OR.fit(Input_OR, target_val)

    print("Final bias and weights of OR gate")

    print(perceptron_OR.W," ",perceptron_OR.b)

   

   # Testing new input

    test=np.array([0,1])

  #  test1=np.insert(test,0,1)

    print("Test value is: ",test)

   #OR gate

    z=perceptron_OR.W.T.dot(test)

    or_out=perceptron_OR.activation_fn(z+perceptron_OR.b)

    print("OR gate prediction: ",or_out)
```
